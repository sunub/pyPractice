# Numpy

```python
import numpy as np

A: list = [[1, 2], [2, 3]]
B: list = [4,5]

print(A + B)

npA = np.array([[1, 2], [3, 4]])
npB = np.array([3, 4])
# broadCast numpy이의 ndarray를 사용하면 이용할 수 있는 기능이다. 2차 배열과 1차 배열을 연산하였을 때 자동으로 2차원에 맞춰 비어있는 배열을 채워 연산하는 기능이다.

print(npA + npB)
# [[4 6],[6 8]]
print(npA * npB)
# [[3, 8], [9, 16]]
print(npA.flatten())
# 2차원 배열을 1차원 배열으로 바꾼다
X = npA.flatten()
X[np.array([0, 1, 2])]
# array에 numpy.array[indexnumbers]를 이용하면 입력한 인덱스의 값들을 배열로 반환한다.
```

# Matplotlib

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
